```python
catdf["Platform"].value_counts()
```

```python
catdf["Genre"].value_counts()
```

```python
sb.catplot(y = "Genre", data = catdf, kind = "count")
plt.show()
```

```python
cPublisher=df['Publisher'].value_counts()
print(cPublisher)
```

```python
cPublisher = cPublisher[cPublisher >= 30]
print(cPublisher)
```

```python
cPublisher.plot(kind='barh')
plt.show()
```

```python
cDeveloper=df['Developer'].value_counts()
print(cDeveloper)
```

```python
cDeveloper = cDeveloper[cDeveloper >= 30]
print(cDeveloper)
```

```python
cDeveloper.plot(kind='barh')
plt.show()
```
