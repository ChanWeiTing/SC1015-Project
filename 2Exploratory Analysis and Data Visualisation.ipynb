### Insights on Categorical Data
#### Publisher and Developer 
Both of these show similarities and we might be able to explore the top developers and publishers based on the number of games they have. We may then be able to classify the top few publishers and developers together to analysis under classification tree models

#### Genre, Rating
These 2 varaibles do not showing any interesting results, though genre can be carefuly analysed to find out the top genres and classify them togther to perform classification.

#### Platform
Platforms allow room for further analysis, we are interested to find out more about how the platforms may affect user score but first we will have to perform a timescaled analysis of the games released by different platforms over the years

#### Name
After checking the raw data, we are able to find out that there are multiple games under the same name but released under different platforms. Example is LEGO Star Wars II: The Original Trilogy which have 8 different games released on PS2, PSP, DS, GC, X360, XB, GBA.

#### User Count and Critic Count
Both variables do not produde well distributed graphs unlike their score counterparts, however these variables may add value when using their related score values, we might use these count values to determine the credibility of the score values in further analysis if needed

#### Year of Release
We analyzed  that our dataset contains mostly data around 2007 as it is harder to obtain complete data of more recent games.
